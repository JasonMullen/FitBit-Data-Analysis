```python
#Check for issues
for name, df in dfs.items():
    print(f"\n **Checking dataset: {name}**")
    
    # Missing values
    missing = df.isnull().sum()
    print(f"Missing values:\n{missing[missing > 0]}\n")
    
    # Duplicates
    duplicate_count = df.duplicated().sum()
    print(f"Total duplicate rows: {duplicate_count}\n")
    
    # Summary statistics
    print("Summary statistics:")
    print(df.describe(), "\n")
```

```python
#Cleaning Process
for name, df in dfs.items():
    for col in df.columns:
        if 'date' in col.lower() or 'time' in col.lower():
            df[col] = pd.to_datetime(df[col], errors='coerce')
```

```python
#Fix Data Types
for name, df in dfs.items():
    for col in df.select_dtypes(include=['object']).columns:
        if df[col].str.isnumeric().all():
            df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
#Remove Duplicates
for name, df in dfs.items():
    dfs[name] = df.drop_duplicates()
```

```python
#Handle Missing Values
for name, df in dfs.items():
    for col in df.columns:
        if df[col].isnull().sum() > 0:
            if df[col].dtype == 'float64' or df[col].dtype == 'int64':  # Numeric columns
                df[col].fillna(df[col].mean(), inplace=True)
            else:  # Categorical columns
                df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
#Standardize Column Names
for name, df in dfs.items():
    df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
#Saving the cleaned data
cleaned_path = "/Users/JumpMan/Downloads/Fitabase Data 3.12.16-4.11.16/new_cleaned_data"
import os
if not os.path.exists(cleaned_path):
    os.makedirs(cleaned_path)

for name, df in dfs.items():
    df.to_csv(f"{cleaned_path}/{name}_cleaned.csv", index=False)
    print(f" {name} dataset saved!")
```

### Total Steps and Calories

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ipywidgets as widgets
from IPython.display import display

# Convert 'activitydate' to datetime for better visualization
dfs['activity_cleaned']['activitydate'] = pd.to_datetime(dfs['activity_cleaned']['activitydate'])

# Get a list of unique IDs plus an option for 'Cumulative'
unique_ids = dfs['activity_cleaned']['id'].unique().tolist()
unique_ids.insert(0, 'Cumulative')

# Dropdown widget for user selection
dropdown = widgets.Dropdown(
    options=unique_ids,
    value='Cumulative',
    description='User ID:',
    style={'description_width': 'initial'}
)

# Function to update plot and display averages with additional calculations
def update_plot(user_id):
    plt.figure(figsize=(12, 5))
    
    if user_id == 'Cumulative':
        # Aggregate data across all users
        df = dfs['activity_cleaned'].groupby('activitydate')[['totalsteps', 'calories', 'veryactiveminutes', 'fairlyactiveminutes', 'lightlyactiveminutes', 'sedentaryminutes']].sum().reset_index()
    else:
        # Filter data for the selected user
        df = dfs['activity_cleaned'][dfs['activity_cleaned']['id'] == user_id]
    
    # Calculate the average of steps and calories burned per day
    avg_steps = df['totalsteps'].mean()
    avg_calories = df['calories'].mean()
    
    # Additional calculations
    total_steps = df['totalsteps'].sum()
    total_calories = df['calories'].sum()
    max_steps = df['totalsteps'].max()
    min_steps = df['totalsteps'].min()
    max_calories = df['calories'].max()
    min_calories = df['calories'].min()
    std_steps = df['totalsteps'].std()
    std_calories = df['calories'].std()

    # Calculate total active minutes
    total_active_minutes = df[['veryactiveminutes', 'fairlyactiveminutes', 'lightlyactiveminutes']].sum().sum()

    # Plot data
    sns.lineplot(data=df, x='activitydate', y='totalsteps', label='Total Steps', marker='o')
    sns.lineplot(data=df, x='activitydate', y='calories', label='Calories Burned', marker='o')

    plt.xticks(rotation=45)
    plt.xlabel("Date")
    plt.ylabel("Count")
    plt.title(f"Daily Steps vs. Calories Burned - {user_id}")
    plt.legend()
    plt.show()

    # Correlation Analysis: Calculate correlation between totalsteps and calories
    correlation_steps_calories = df['totalsteps'].corr(df['calories'])
    correlation_steps_active_minutes = df['totalsteps'].corr(df[['veryactiveminutes', 'fairlyactiveminutes', 'lightlyactiveminutes']].sum(axis=1))
    correlation_calories_active_minutes = df['calories'].corr(df[['veryactiveminutes', 'fairlyactiveminutes', 'lightlyactiveminutes']].sum(axis=1))
    correlation_sedentary_minutes_steps = df['sedentaryminutes'].corr(df['totalsteps'])
    correlation_sedentary_minutes_calories = df['sedentaryminutes'].corr(df['calories'])

    # Display the averages and other calculations
    print(f"Average Steps per Day: {avg_steps:.2f}")
    print(f"Average Calories Burned per Day: {avg_calories:.2f}")
    print(f"Total Steps: {total_steps}")
    print(f"Total Calories Burned: {total_calories}")
    print(f"Max Steps: {max_steps}")
    print(f"Min Steps: {min_steps}")
    print(f"Max Calories Burned: {max_calories}")
    print(f"Min Calories Burned: {min_calories}")
    print(f"Standard Deviation of Steps: {std_steps:.2f}")
    print(f"Standard Deviation of Calories: {std_calories:.2f}")
    print(f"Total Active Minutes (Very, Fairly, and Lightly Active): {total_active_minutes} minutes")
    
    # Display the correlation results
    print(f"\nCorrelation between Total Steps and Calories Burned: {correlation_steps_calories:.2f}")
    print(f"Correlation between Total Steps and Active Minutes: {correlation_steps_active_minutes:.2f}")
    print(f"Correlation between Calories and Active Minutes: {correlation_calories_active_minutes:.2f}")
    print(f"Correlation between Sedentary Minutes and Total Steps: {correlation_sedentary_minutes_steps:.2f}")
    print(f"Correlation between Sedentary Minutes and Calories: {correlation_sedentary_minutes_calories:.2f}")
    
    # Additional Graphs

    # 1. Histogram of Total Steps and Calories
    plt.figure(figsize=(12, 5))
    sns.histplot(df['totalsteps'], bins=30, kde=True, color='blue', label='Total Steps', alpha=0.6)
    sns.histplot(df['calories'], bins=30, kde=True, color='red', label='Calories Burned', alpha=0.6)
    plt.title('Histogram of Total Steps and Calories Burned')
    plt.xlabel('Count')
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()

    # 2. Boxplot of Active Minutes (Very, Fairly, and Lightly Active)
    plt.figure(figsize=(12, 5))
    sns.boxplot(data=df[['veryactiveminutes', 'fairlyactiveminutes', 'lightlyactiveminutes']])
    plt.title('Boxplot of Active Minutes')
    plt.xlabel('Activity Type')
    plt.ylabel('Minutes')
    plt.show()

    # 3. Rolling Average of Total Steps and Calories
    df['rolling_steps'] = df['totalsteps'].rolling(window=7).mean()
    df['rolling_calories'] = df['calories'].rolling(window=7).mean()
    
    plt.figure(figsize=(12, 5))
    sns.lineplot(data=df, x='activitydate', y='rolling_steps', label='7-Day Rolling Average - Steps', color='blue')
    sns.lineplot(data=df, x='activitydate', y='rolling_calories', label='7-Day Rolling Average - Calories', color='red')
    plt.xticks(rotation=45)
    plt.xlabel("Date")
    plt.ylabel("Rolling Average")
    plt.title(f"7-Day Rolling Average of Steps and Calories - {user_id}")
    plt.legend()
    plt.show()

    # 4. Correlation Heatmap
    correlation_matrix = df[['totalsteps', 'calories', 'veryactiveminutes', 'fairlyactiveminutes', 'lightlyactiveminutes', 'sedentaryminutes']].corr()
    
    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
    plt.title('Correlation Heatmap')
    plt.show()

# Display dropdown and connect it to the update_plot function
widgets.interactive(update_plot, user_id=dropdown)
```

### HeartRate and BPM

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ipywidgets as widgets
from IPython.display import display

# Convert 'time' to datetime for better visualization
dfs['heart_rate_cleaned']['time'] = pd.to_datetime(dfs['heart_rate_cleaned']['time'])

# Get a list of unique IDs plus an option for 'Cumulative'
unique_ids = dfs['heart_rate_cleaned']['id'].unique().tolist()
unique_ids.insert(0, 'Cumulative')

# Dropdown widget for user selection
dropdown = widgets.Dropdown(
    options=unique_ids,
    value='Cumulative',
    description='User ID:',
    style={'description_width': 'initial'}
)

# Function to update plot and display averages with additional calculations
def update_plot(user_id):
    plt.figure(figsize=(12, 5))
    
    if user_id == 'Cumulative':
        # Aggregate data across all users
        df = dfs['heart_rate_cleaned'].groupby('time')[['value']].mean().reset_index()
    else:
        # Filter data for the selected user
        df = dfs['heart_rate_cleaned'][dfs['heart_rate_cleaned']['id'] == user_id]
    
    # Calculate the average heart rate
    avg_heart_rate = df['value'].mean()
    
    # Additional calculations
    max_heart_rate = df['value'].max()
    min_heart_rate = df['value'].min()
    std_heart_rate = df['value'].std()
    
    # Peak heart rate times
    peak_heart_rate_time = df.loc[df['value'].idxmax()]['time']
    
    # Plot data for heart rate
    sns.lineplot(data=df, x='time', y='value', color='red', label='Heart Rate (BPM)', marker='o')
    plt.xticks(rotation=45)
    plt.xlabel("Time")
    plt.ylabel("Heart Rate (BPM)")
    plt.title(f"Heart Rate Trends Over Time - {user_id}")
    plt.legend()
    plt.show()

    # Display the calculations
    print(f"Average Heart Rate: {avg_heart_rate:.2f} BPM")
    print(f"Max Heart Rate: {max_heart_rate} BPM")
    print(f"Min Heart Rate: {min_heart_rate} BPM")
    print(f"Standard Deviation of Heart Rate: {std_heart_rate:.2f} BPM")
    print(f"Peak Heart Rate occurred at: {peak_heart_rate_time}")
    
    # Additional Graphs

    # 1. Histogram of Heart Rate
    plt.figure(figsize=(12, 5))
    sns.histplot(df['value'], bins=30, kde=True, color='blue', label='Heart Rate Distribution', alpha=0.6)
    plt.title('Histogram of Heart Rate')
    plt.xlabel('Heart Rate (BPM)')
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()

    # 2. Boxplot of Heart Rate
    plt.figure(figsize=(12, 5))
    sns.boxplot(data=df['value'], color='green')
    plt.title('Boxplot of Heart Rate')
    plt.ylabel('Heart Rate (BPM)')
    plt.show()

    # 3. Rolling Average of Heart Rate (7-day rolling average)
    df['rolling_heart_rate'] = df['value'].rolling(window=7).mean()
    
    plt.figure(figsize=(12, 5))
    sns.lineplot(data=df, x='time', y='rolling_heart_rate', color='purple', label='7-Day Rolling Average Heart Rate')
    plt.xticks(rotation=45)
    plt.xlabel("Time")
    plt.ylabel("Rolling Average Heart Rate (BPM)")
    plt.title(f"7-Day Rolling Average of Heart Rate - {user_id}")
    plt.legend()
    plt.show()

    # 4. Correlation Heatmap with Heart Rate and Other Variables
    if 'activity_cleaned' in dfs and 'minute_steps_cleaned' in dfs:
        # Check columns in activity_cleaned and minute_steps_cleaned
        print(f"Columns in activity_cleaned: {dfs['activity_cleaned'].columns}")
        print(f"Columns in minute_steps_cleaned: {dfs['minute_steps_cleaned'].columns}")

        # Assuming 'activitydate' is the equivalent column in activity_cleaned for time, we will merge on that
        if 'activitydate' in dfs['activity_cleaned'].columns and 'time' in df.columns:
            merged_df = pd.merge(df, dfs['activity_cleaned'], left_on='time', right_on='activitydate', how='inner')
            merged_df = pd.merge(merged_df, dfs['minute_steps_cleaned'], left_on='time', right_on='time', how='inner')

            # Calculate the correlation matrix between heart rate, steps, and calories
            correlation_matrix = merged_df[['value', 'totalsteps', 'calories', 'veryactiveminutes', 'fairlyactiveminutes', 'lightlyactiveminutes']].corr()
            
            # Plot correlation heatmap
            plt.figure(figsize=(10, 8))
            sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
            plt.title('Correlation Heatmap: Heart Rate vs. Steps, Calories, and Active Minutes')
            plt.show()

    # 5. Pearson Correlation Coefficients (Numerical Output)
    if 'activity_cleaned' in dfs and 'minute_steps_cleaned' in dfs:
        # Correlation between heart rate and total steps
        steps_correlation = df['value'].corr(dfs['minute_steps_cleaned']['value'])
        print(f"Pearson Correlation between Heart Rate and Total Steps: {steps_correlation:.2f}")

        # Correlation between heart rate and calories
        calories_correlation = df['value'].corr(dfs['activity_cleaned']['calories'])
        print(f"Pearson Correlation between Heart Rate and Calories Burned: {calories_correlation:.2f}")

        # Correlation between heart rate and active minutes
        active_minutes_correlation = df['value'].corr(dfs['activity_cleaned'][['veryactiveminutes', 'fairlyactiveminutes', 'lightlyactiveminutes']].sum(axis=1))
        print(f"Pearson Correlation between Heart Rate and Active Minutes: {active_minutes_correlation:.2f}")

# Display dropdown and connect it to the update_plot function
widgets.interactive(update_plot, user_id=dropdown)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ipywidgets as widgets
from IPython.display import display

# Convert 'activitydate' to datetime for better visualization
dfs['activity_cleaned']['activitydate'] = pd.to_datetime(dfs['activity_cleaned']['activitydate'])

# Get a list of unique IDs plus an option for 'Cumulative'
unique_ids = dfs['activity_cleaned']['id'].unique().tolist()
unique_ids.insert(0, 'Cumulative')

# Dropdown widget for user selection
dropdown = widgets.Dropdown(
    options=unique_ids,
    value='Cumulative',
    description='User ID:',
    style={'description_width': 'initial'}
)

# Function to update plot and display average activity time with additional calculations
def update_plot(user_id):
    plt.figure(figsize=(12, 5))
    
    if user_id == 'Cumulative':
        # Aggregate data across all users
        df = dfs['activity_cleaned'].groupby('activitydate')[['veryactiveminutes', 'fairlyactiveminutes', 'lightlyactiveminutes', 'sedentaryminutes', 'totalsteps', 'calories']].sum().reset_index()
    else:
        # Filter data for the selected user
        df = dfs['activity_cleaned'][dfs['activity_cleaned']['id'] == user_id]
    
    # Plot the stackplot for activity levels
    plt.stackplot(
        df['activitydate'],
        df['veryactiveminutes'],
        df['fairlyactiveminutes'],
        df['lightlyactiveminutes'],
        df['sedentaryminutes'],
        labels=["Very Active", "Fairly Active", "Lightly Active", "Sedentary"],
        alpha=0.7
    )

    plt.xticks(rotation=45)
    plt.xlabel("Date")
    plt.ylabel("Minutes")
    plt.title(f"Breakdown of Activity Levels Over Time - {user_id}")
    plt.legend()
    plt.show()

    # Calculate the average time spent in each activity level
    avg_very_active = df['veryactiveminutes'].mean()
    avg_fairly_active = df['fairlyactiveminutes'].mean()
    avg_lightly_active = df['lightlyactiveminutes'].mean()
    avg_sedentary = df['sedentaryminutes'].mean()

    # Calculate the total time spent in activity
    total_time = avg_very_active + avg_fairly_active + avg_lightly_active + avg_sedentary

    # Calculate the percentage of time spent in each activity level
    pct_very_active = (avg_very_active / total_time) * 100
    pct_fairly_active = (avg_fairly_active / total_time) * 100
    pct_lightly_active = (avg_lightly_active / total_time) * 100
    pct_sedentary = (avg_sedentary / total_time) * 100

    # Max and Min Time in Each Activity Level
    max_very_active_day = df.loc[df['veryactiveminutes'].idxmax()]
    max_fairly_active_day = df.loc[df['fairlyactiveminutes'].idxmax()]
    max_lightly_active_day = df.loc[df['lightlyactiveminutes'].idxmax()]
    max_sedentary_day = df.loc[df['sedentaryminutes'].idxmax()]

    # Convert minutes to hours for display (1 hour = 60 minutes)
    avg_very_active_hours = avg_very_active / 60
    avg_fairly_active_hours = avg_fairly_active / 60
    avg_lightly_active_hours = avg_lightly_active / 60
    avg_sedentary_hours = avg_sedentary / 60

    # Display the averages in both hours and minutes
    print(f"Average Time Spent (minutes and hours):")
    print(f"  Very Active: {avg_very_active_hours:.2f} hours / {avg_very_active:.2f} minutes")
    print(f"  Fairly Active: {avg_fairly_active_hours:.2f} hours / {avg_fairly_active:.2f} minutes")
    print(f"  Lightly Active: {avg_lightly_active_hours:.2f} hours / {avg_lightly_active:.2f} minutes")
    print(f"  Sedentary: {avg_sedentary_hours:.2f} hours / {avg_sedentary:.2f} minutes")

    # Display additional calculations
    print(f"\nTotal Time Spent Across All Activity Levels: {total_time:.2f} minutes / {total_time / 60:.2f} hours")
    print(f"\nPercentage of Time Spent in Each Activity Level:")
    print(f"  Very Active: {pct_very_active:.2f}%")
    print(f"  Fairly Active: {pct_fairly_active:.2f}%")
    print(f"  Lightly Active: {pct_lightly_active:.2f}%")
    print(f"  Sedentary: {pct_sedentary:.2f}%")

    print(f"\nMax Time Spent on Each Activity Level:")
    print(f"  Max Very Active: {max_very_active_day['activitydate']} - {max_very_active_day['veryactiveminutes']} minutes")
    print(f"  Max Fairly Active: {max_fairly_active_day['activitydate']} - {max_fairly_active_day['fairlyactiveminutes']} minutes")
    print(f"  Max Lightly Active: {max_lightly_active_day['activitydate']} - {max_lightly_active_day['lightlyactiveminutes']} minutes")
    print(f"  Max Sedentary: {max_sedentary_day['activitydate']} - {max_sedentary_day['sedentaryminutes']} minutes")

# Display dropdown and connect it to the update_plot function
widgets.interactive(update_plot, user_id=dropdown)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ipywidgets as widgets
from IPython.display import display
from scipy.stats import skew, kurtosis

# Define file paths for all the datasets
file_paths = {
    'heart_rate_cleaned': '/Users/JumpMan/Downloads/Fitabase Data 3.12.16-4.11.16/cleaned_data/heart_rate_cleaned.csv',
    'hourly_calories_cleaned': '/Users/JumpMan/Downloads/Fitabase Data 3.12.16-4.11.16/cleaned_data/hourly_calories_cleaned.csv',
    'hourly_intensity_cleaned': '/Users/JumpMan/Downloads/Fitabase Data 3.12.16-4.11.16/cleaned_data/hourly_intensity_cleaned.csv',
    'hourly_steps_cleaned': '/Users/JumpMan/Downloads/Fitabase Data 3.12.16-4.11.16/cleaned_data/hourly_steps_cleaned.csv',
    'minute_calories_cleaned': '/Users/JumpMan/Downloads/Fitabase Data 3.12.16-4.11.16/cleaned_data/minute_calories_cleaned.csv',
    'minute_intensity_cleaned': '/Users/JumpMan/Downloads/Fitabase Data 3.12.16-4.11.16/cleaned_data/minute_intensity_cleaned.csv',
    'minute_sleep_cleaned': '/Users/JumpMan/Downloads/Fitabase Data 3.12.16-4.11.16/cleaned_data/minute_sleep_cleaned.csv',
    'minute_met_cleaned': '/Users/JumpMan/Downloads/Fitabase Data 3.12.16-4.11.16/cleaned_data/minute_met_cleaned.csv',
    'minute_steps_cleaned': '/Users/JumpMan/Downloads/Fitabase Data 3.12.16-4.11.16/cleaned_data/minute_steps_cleaned.csv',
    'weight_log_cleaned': '/Users/JumpMan/Downloads/Fitabase Data 3.12.16-4.11.16/cleaned_data/weight_log_cleaned.csv',
    'activity_cleaned': '/Users/JumpMan/Downloads/Fitabase Data 3.12.16-4.11.16/cleaned_data/activity_cleaned.csv'
}

# Deeper metrics calculation with handling for precision loss
def calculate_deeper_metrics(df, column):
    median = df[column].median()
    iqr = df[column].quantile(0.75) - df[column].quantile(0.25)
    
    # Check for precision loss before calculating skewness and kurtosis
    if df[column].std() > 1e-5:  # Check if standard deviation is larger than a small threshold
        skewness = skew(df[column].dropna())  # Drop NaN values before calculating skew
        kurt = kurtosis(df[column].dropna())  # Drop NaN values before calculating kurtosis
    else:
        skewness = None
        kurt = None

    return {
        'median': median,
        'IQR': iqr,
        'skewness': skewness,
        'kurtosis': kurt
    }

# Function to calculate metrics
def calculate_metrics(df, column):
    total_change = df[column].iloc[-1] - df[column].iloc[0]
    avg_value = df[column].mean()
    max_value = df[column].max()
    min_value = df[column].min()
    std_dev = df[column].std()
    variance = df[column].var()
    
    return {
        'total_change': total_change,
        'average': avg_value,
        'max': max_value,
        'min': min_value,
        'std_dev': std_dev,
        'variance': variance
    }

# Function to process datasets
def process_datasets():
    user_metrics = []
    for dataset_name, file_path in file_paths.items():
        df = pd.read_csv(file_path)
        
        # Adjust column name dynamically if needed
        value_column = None
        if 'heart_rate' in df.columns:
            value_column = 'heart_rate'
        elif 'steps' in df.columns:
            value_column = 'steps'
        elif 'calories' in df.columns:
            value_column = 'calories'
        elif 'intensity' in df.columns:
            value_column = 'intensity'
        else:
            value_column = df.columns[0]  # Fallback to first column if no specific match

        unique_ids = df['id'].unique().tolist()
        unique_ids.insert(0, 'Cumulative')  # Insert 'Cumulative' option
        
        for user_id in unique_ids[1:]:
            user_data = df[df['id'] == user_id] if user_id != 'Cumulative' else df
            metrics = calculate_metrics(user_data, value_column)
            deeper_metrics = calculate_deeper_metrics(user_data, value_column)  # Additional metrics
            metrics.update(deeper_metrics)
            
            metrics['user_id'] = user_id
            metrics['dataset'] = dataset_name
            
            user_metrics.append(metrics)

    metrics_df = pd.DataFrame(user_metrics)
    return metrics_df

# Process the datasets and calculate the metrics
metrics_df = process_datasets()

# Centralize the data
centralized_df = metrics_df.copy()

for col in ['total_change', 'average', 'max', 'min', 'std_dev', 'variance', 'median', 'IQR', 'skewness', 'kurtosis']:
    centralized_df[col] = (metrics_df[col] - metrics_df[col].mean()) / metrics_df[col].std()

# Display centralized data
display(centralized_df)

# Visualizations
plt.figure(figsize=(12, 6))
sns.barplot(x='user_id', y='total_change', hue='dataset', data=centralized_df, palette='Set2')
plt.title('Centralized Total Change Across Users and Datasets')
plt.xlabel('User ID')
plt.ylabel('Centralized Total Change')
plt.xticks(rotation=45)
plt.legend(title='Dataset', loc='upper right')
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(x='user_id', y='median', hue='dataset', data=centralized_df, palette='Set1')
plt.title('Median Value Across Users and Datasets')
plt.xlabel('User ID')
plt.ylabel('Median')
plt.xticks(rotation=45)
plt.legend(title='Dataset', loc='upper right')
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(x='user_id', y='IQR', hue='dataset', data=centralized_df, palette='Set2')
plt.title('IQR Across Users and Datasets')
plt.xlabel('User ID')
plt.ylabel('Interquartile Range (IQR)')
plt.xticks(rotation=45)
plt.legend(title='Dataset', loc='upper right')
plt.show()

# Add an interactive widget to search user_id
def search_user_data(user_id):
    filtered_df = centralized_df[centralized_df['user_id'] == user_id]
    display(filtered_df)

user_id_widget = widgets.Dropdown(
    options=centralized_df['user_id'].unique(),
    description='User ID:',
    disabled=False
)

widgets.interactive(search_user_data, user_id=user_id_widget)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Load cleaned dataset
file_path = '/Users/JumpMan/Downloads/Fitabase Data 3.12.16-4.11.16/cleaned_data/activity_cleaned.csv'
df = pd.read_csv(file_path)

# Selecting features and target
features = ['totalsteps', 'veryactiveminutes', 'fairlyactiveminutes', 'lightlyactiveminutes', 'sedentaryminutes']
target = 'calories'

# Drop rows with missing values
df = df.dropna(subset=features + [target])

# Handle outliers using IQR method
def remove_outliers(df, cols):
    for col in cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    return df

df = remove_outliers(df, features + [target])

# Splitting data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(df[features], df[target], test_size=0.2, random_state=42)

# Hyperparameter tuning for Random Forest using GridSearchCV
rf_params = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}
rf_grid_search = GridSearchCV(RandomForestRegressor(random_state=42), rf_params, cv=3, scoring='neg_mean_absolute_error', n_jobs=-1)
rf_grid_search.fit(X_train, y_train)
best_rf_model = rf_grid_search.best_estimator_

# Hyperparameter tuning for XGBoost using GridSearchCV
xgb_params = {
    'n_estimators': [100, 200],
    'learning_rate': [0.05, 0.1],
    'max_depth': [3, 5],
    'subsample': [0.8, 1],
    'colsample_bytree': [0.8, 1]
}
xgb_grid_search = GridSearchCV(XGBRegressor(random_state=42), xgb_params, cv=3, scoring='neg_mean_absolute_error', n_jobs=-1)
xgb_grid_search.fit(X_train, y_train)
best_xgb_model = xgb_grid_search.best_estimator_

# Predictions with tuned models
rf_preds = best_rf_model.predict(X_test)
xgb_preds = best_xgb_model.predict(X_test)

# Evaluation Metrics
rf_mae = mean_absolute_error(y_test, rf_preds)
rf_rmse = np.sqrt(mean_squared_error(y_test, rf_preds))

xgb_mae = mean_absolute_error(y_test, xgb_preds)
xgb_rmse = np.sqrt(mean_squared_error(y_test, xgb_preds))

print(f"Random Forest - MAE: {rf_mae:.2f}, RMSE: {rf_rmse:.2f}")
print(f"XGBoost - MAE: {xgb_mae:.2f}, RMSE: {xgb_rmse:.2f}")

# Feature Importance
rf_feature_importances = pd.Series(best_rf_model.feature_importances_, index=features).sort_values(ascending=False)
xgb_feature_importances = pd.Series(best_xgb_model.feature_importances_, index=features).sort_values(ascending=False)

print("\nRandom Forest Feature Importances:")
print(rf_feature_importances)

print("\nXGBoost Feature Importances:")
print(xgb_feature_importances)

# Visualization: Pie Charts for Feature Importance
plt.figure(figsize=(12, 6))

# Random Forest Pie Chart
plt.subplot(1, 2, 1)
plt.pie(rf_feature_importances.values,
        labels=rf_feature_importances.index,
        autopct='%1.1f%%',
        startangle=140,
        colors=plt.cm.Paired.colors)
plt.title('Random Forest Feature Importances')

# XGBoost Pie Chart
plt.subplot(1, 2, 2)
plt.pie(xgb_feature_importances.values,
        labels=xgb_feature_importances.index,
        autopct='%1.1f%%',
        startangle=140,
        colors=plt.cm.Paired.colors)
plt.title('XGBoost Feature Importances')

plt.tight_layout()
plt.show()

# Visualization: Bar Chart for Model Performance Metrics
metrics = ['MAE', 'RMSE']
rf_values = [rf_mae, rf_rmse]
xgb_values = [xgb_mae, xgb_rmse]

x = np.arange(len(metrics))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(8, 6))
rects1 = ax.bar(x - width/2, rf_values, width, label='Random Forest', color='skyblue')
rects2 = ax.bar(x + width/2, xgb_values, width, label='XGBoost', color='orange')

# Add labels and title
ax.set_ylabel('Error')
ax.set_title('Model Performance Comparison')
ax.set_xticks(x)
ax.set_xticklabels(metrics)
ax.legend()

# Add value annotations to bars
for rect in rects1 + rects2:
    height = rect.get_height()
    ax.annotate(f'{height:.2f}',
                xy=(rect.get_x() + rect.get_width() / 2, height),
                xytext=(0, 3),  # offset text above bar
                textcoords="offset points",
                ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

## 5. Reinforcement Learning

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# File Paths
file_paths = {
    'heart_rate_cleaned': '/Users/JumpMan/Downloads/Fitabase Data 3.12.16-4.11.16/cleaned_data/heart_rate_cleaned.csv',
    'hourly_calories_cleaned': '/Users/JumpMan/Downloads/Fitabase Data 3.12.16-4.11.16/cleaned_data/hourly_calories_cleaned.csv',
    'hourly_intensity_cleaned': '/Users/JumpMan/Downloads/Fitabase Data 3.12.16-4.11.16/cleaned_data/hourly_intensity_cleaned.csv',
    'hourly_steps_cleaned': '/Users/JumpMan/Downloads/Fitabase Data 3.12.16-4.11.16/cleaned_data/hourly_steps_cleaned.csv'
}

# Load the data
dfs = {}
for name, path in file_paths.items():
    dfs[name] = pd.read_csv(path)

# Merge datasets based on 'id'
merged_df = None
for key, df in dfs.items():
    if 'id' in df.columns:
        if merged_df is None:
            merged_df = df
        else:
            merged_df = pd.merge(merged_df, df, on='id', how='outer')

# Drop non-numeric columns except 'id'
numeric_columns = merged_df.select_dtypes(include=['number']).columns.tolist()
numeric_columns.remove('id')  # Keep 'id' separate

# Handle missing values by filling with median
for col in numeric_columns:
    merged_df[col].fillna(merged_df[col].median(), inplace=True)

# Normalize the numeric columns
scaler = StandardScaler()
merged_df[numeric_columns] = scaler.fit_transform(merged_df[numeric_columns])

# Let's focus on a few columns for simplicity (e.g., 'steps', 'calories', 'intensity')
data = merged_df[['hourly_steps_cleaned', 'hourly_calories_cleaned', 'hourly_intensity_cleaned']]

# Q-learning parameters
state_space = 5  # e.g., 5 discrete states based on the activity level (rest, light, moderate, high, max)
action_space = 4  # e.g., 4 possible actions (0 - No activity, 1 - Light, 2 - Moderate, 3 - Intense)
q_table = np.zeros((state_space, action_space))  # Initialize Q-table

learning_rate = 0.1
discount_factor = 0.95
epsilon = 0.1  # Exploration rate

# Define reward function (here we use calories burned as reward)
def reward_function(state, action, steps, calories, intensity):
    if action == 0:  # No activity
        return -0.5 * calories  # Penalty for no activity
    elif action == 1:  # Light activity
        return 0.2 * calories
    elif action == 2:  # Moderate activity
        return 0.5 * calories + 0.1 * steps
    elif action == 3:  # Intense activity
        return 1.0 * calories + 0.2 * steps
    return 0

# Training Q-learning agent
episodes = 1000

for episode in range(episodes):
    # Randomly select a state (e.g., current step intensity)
    current_state = np.random.randint(0, state_space)
    
    # Choose an action using epsilon-greedy strategy
    if np.random.uniform(0, 1) < epsilon:
        action = np.random.randint(0, action_space)  # Exploration
    else:
        action = np.argmax(q_table[current_state])  # Exploitation
    
    # Simulate taking action and transitioning to the next state
    steps = np.random.choice(data['hourly_steps_cleaned'])
    calories = np.random.choice(data['hourly_calories_cleaned'])
    intensity = np.random.choice(data['hourly_intensity_cleaned'])
    
    # Get reward based on action taken
    reward = reward_function(current_state, action, steps, calories, intensity)
    
    # Update Q-table using Q-learning update rule
    next_state = np.random.randint(0, state_space)  # For simplicity, we randomly transition to a new state
    best_future_q = np.max(q_table[next_state])
    q_table[current_state, action] = (1 - learning_rate) * q_table[current_state, action] + \
                                      learning_rate * (reward + discount_factor * best_future_q)

# After training, print Q-table
print("Q-table after training:")
print(q_table)

# Visualizing the Q-table
plt.imshow(q_table, cmap='hot', interpolation='nearest')
plt.colorbar()
plt.title('Q-table Heatmap')
plt.xlabel('Actions')
plt.ylabel('States')
plt.show()

```
